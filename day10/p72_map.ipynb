- 지도에 마커표시
    - Maker() 클래스로 생성한 뒤 지도에 추가
    - Icon() 내 들어가는 icon 속성값은 https://getbootstrap.com/docs/3.3/components/ 에서 확인할것

```python
#마커
map = folium.Map(location=[35.178117, 129.198921], tiles='CartoDB.Voyager' ,zoom_start=16)

folium.Marker(location=[35.178117, 129.198921], tooltip='송정 해수욕장', popup='서핑하러 가즈아').add_to(map)
popup= folium.Popup('<pre><a href="https://www.starbucks.co.kr/">스타벅스</a>로 오니라~</pre>', min_width=200, max_width=300)
folium.Marker(location=[35.180762, 129.202324], tooltip='스타벅스 부산송정비치점', popup=popup, icon=folium.Icon(color='green', icon='heart')).add_to(map)

map
```
